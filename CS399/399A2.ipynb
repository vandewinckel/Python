```python
csv = open('players_20.csv',encoding='utf8')
csvL = []
csvLs = []
playr = []

overBMI = []
underBMI = []

over32 = []

for line in csv:
    csvL.append(line)
for players in csvL:
    x = players.split(',')
    csvLs.append(x)
del csvLs[0]
for players_dat in csvLs:
    #[2]name [4]age [6]height [7]weight [8]nationality [9]club [14]positions 
    #print(players_dat[2], players_dat[4], players_dat[6], players_dat[7], players_dat[8], players_dat[9], players_dat[14])
    playr.append([players_dat[9], players_dat[2],"{0:.2f}".format(float(players_dat[7])/((float(players_dat[6]))*.01)**2), players_dat[4]])
for plr in playr:
    if float(plr[2]) > 24.9:
        #print("OVER 24.9", plr[1], 'BMI:', plr[2])
        overBMI.append(plr[1])
    if float(plr[2]) < 18.4:
        #print("UNDER 18.4",(plr[1]), 'BMI:', plr[2])
        underBMI.append(plr[1])
    if int(plr[3]) > 32:
        #print(plr[1],plr[3])
        over32.append(plr[1])
```
